```python
df = pd.read_csv ('/content/drive/MyDrive/RA/entity_activity_table_clusters.csv')
df
```

```python
#Listing down the unique dates in the dataset
lst_of_dates = list(df['SENT_AT_MONTH'].value_counts().index)
lst_of_dates
```

```python

#Created a for loop where the topical data is assigned to every date of the month
#Handle is the index of the dataframe
for date in lst_of_dates:
  temp_df = df[['HANDLE','mean_topic1', 'mean_topic2', 'mean_topic3', 'mean_topic4','mean_topic5']].loc[df['SENT_AT_MONTH'] == date]
  temp_df.set_index('HANDLE', inplace = True)
  temp_df = temp_df.apply(pd.to_numeric)
  temp_df = temp_df.T
  corr_mat = temp_df.corr()
  corr_mat.fillna('NA', inplace = True)
  corr_mat.to_csv(date + '.csv')
```
